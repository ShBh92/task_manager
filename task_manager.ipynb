```python
# Lets start coding the project!
# 1. User Authentication:
# • Registration:
# o Create a function to prompt the user to enter a username and
# password
# o Ensure that the username is unique, and hash the password for
# security before storing it in a file
# • Login:
# o Create a function to prompt the user for their username and
# password, validate the credentials by comparing them with the stored
# data, and grant access to the task manager upon successful login

# Function for user registration
import hashlib
def register():
  username = input("Enter a username: ")
  password = input("Enter a password: ")

  # Hash the password for security
  hashed_password = hashlib.sha256(password.encode()).hexdigest()

  # Check if user is already registered user
  # add exception handling
  try:
    with open("users.txt", "r") as file:
      for line in file:
        stored_username = line.split(",")[0]
        if username == stored_username:
          print("Username already exists. Please choose a different one.")
          return
        # Store the username and hashed password in a file
    with open("users.txt", "a") as file:
      file.write(f"{username},{hashed_password}\n")
      print("Registration successful!")
  except FileNotFoundError:
      with open("users.txt", "w") as file:
        file.write(f"{username},{hashed_password}\n")
        print("Registration successful!")
  except Exception as e:
    print(f"An error occurred during registration: {e}")
```

```python
# Function for user login
def login():
  username = input("Enter your username: ")
  password = input("Enter your password: ")
  hashed_password = hashlib.sha256(password.encode()).hexdigest()

  try:
    with open("users.txt", "r") as file:
      for line in file:
        stored_username, stored_hashed_password = line.strip().split(",")
        if username == stored_username and hashed_password == stored_hashed_password:
          print("Login successful!")
          return username

    print("Invalid username or password.")
    return None
  except FileNotFoundError:
    print("User not found. Please register first.")
    return None
#
```

```python
import os

# Function to add a new task
def add_task(user):
  task_description = input("Enter the task description: ")

  user_task_file = f'{user}_tasks.txt'
  tasks = []
  next_task_id = 0

  # Read existing tasks and determine the next sequential ID
  if os.path.exists(user_task_file):
    with open(user_task_file, 'r') as file:
      for line in file:
        line_stripped = line.strip()
        if line_stripped: # Only process non-empty lines
          parts = line_stripped.split(',', 2)
          if len(parts) == 3: # Ensure line is correctly formatted
            current_id = int(parts[0])
            tasks.append({'id': parts[0], 'description': parts[1], 'status': parts[2]})
            if current_id >= next_task_id:
              next_task_id = current_id + 1
          else:
            print(f"Warning: Skipping malformed line in task file: {line_stripped}")

  # Add the new task
  new_task_id = next_task_id
  tasks.append({'id': str(new_task_id), 'description': task_description, 'status': 'Pending'})

  # Write all tasks back to the file
  with open(user_task_file, 'w') as file: # Use 'w' mode to overwrite the file
    for task in tasks:
      file.write(f"{task['id']},{task['description']},{task['status']}\n")
  print("Task added successfully!")
```

```python
# Function to view tasks
def view_tasks(user):
  print("\nYour tasks:")
  try:
    with open(f'{user}_tasks.txt', 'r') as file:
      tasks_found = False
      for line in file:
        # Expecting format: task_id,task_description,status
        task_id, task_description, status = line.strip().split(',', 2)
        print(f"ID: {task_id}, Task: {task_description}, Status: {status}")
        tasks_found = True
      if not tasks_found:
        print("No tasks found. Please add a task first.")
  except FileNotFoundError:
    print("No tasks file found for this user. Please add a task first.")
  except ValueError:
    print("Error: Tasks are not in the expected format (ID, Description, Status). Please ensure 'add_task' is updated to include task IDs.")
  except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# Function to mark a task as completed
def mark_task_as_completed(username):
  task_id_to_mark = input("Enter the task ID to mark as completed: ")
  tasks = []
  updated = False
  try:
    with open(f"{username}_tasks.txt", "r") as file:
      for line in file:
        task_id, description, status = line.strip().split(',', 2)
        tasks.append({'id': task_id, 'description': description, 'status': status})

    with open(f"{username}_tasks.txt", "w") as file:
      for task in tasks:
        if task['id'] == task_id_to_mark:
          task['status'] = "Completed"
          updated = True
        file.write(f"{task['id']},{task['description']},{task['status']}\n")

    if updated:
      print(f"Task ID {task_id_to_mark} marked as completed!")
    else:
      print(f"Task with ID {task_id_to_mark} not found.")
  except FileNotFoundError:
    print("No tasks found for this user.")
  except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Function to delete a task
def delete_task(username):
  print(f"In delete ******")
  task_id = input("Enter the task ID to delete: ")
  try:
    tasks = []
    with open(f"{username}_tasks.txt", "r") as file:
      tasks = [line.strip().split(',') for line in file]

      for task in tasks:
        if task[0] == task_id:
          print(f"task id to remove ==== {task[0]}")
          tasks.remove(task)
          print(f"Task {task_id} removed successfully")
          with open(f"{username}_tasks.txt", "w") as file:
            print(f"new task ==== {tasks}")
            if len(tasks) == 0:
                print(f"{username}_tasks.txt file deleted successfully as there are no pending tasks left for user {username} ")
                # remove file
                os.remove(f"{username}_tasks.txt")
            for task in tasks:
              file.write(f"{task[0]},{task[1]},{task[2]}\n")


  except FileNotFoundError:
    print("No tasks found for this user.")
    return
  except Exception as e:
    print(f"An error occurred: {e}")
    return
```

```python
# Main menu after user logs in
def task_manager_menu(username):
  while True:
    print("\nTask Manager Menu:")
    print("1. View Task")
    print("2. Add Task")
    print("3. Mark Task as Completed")
    print("4. Delete Task")
    print("5. Logout")
    choice = input("Enter your choice: ")
    if choice == "1":
      view_tasks(username)
    elif choice == "2":
      print("Adding...")
      add_task(username)
    elif choice == "3":
      mark_task_as_completed(username)
    elif choice == "4":
      delete_task(username)
    elif choice == "5":
      print("Logging out...")
      break
    else:
      print("Invalid choice. Please try again.")
```

```python
# Main program flow for registration and login
def main():
  while True:
    print("\n 1.Register")
    print("2. Login")
    print("3. Exit")
    choice = input("Enter your choice: ")
    if choice == "1":
      register()
    elif choice == "2":
      username = login()
      if username:
        task_manager_menu(username)
    elif choice == "3":
      print("Exiting the program.")
      break
    else:

      print("Invalid choice. Please try again.")
#Lets execute the main function to start the task manager
if __name__ == "__main__":
  main()

```
